```python
import cv2

import matplotlib.pyplot as plt
```

```python
image_paths = [r'C:\Users\student\Downloads\ulb1.jpg',
               
               r'C:\Users\student\Downloads\ulb2.jpg',
               
               r'C:\Users\student\Downloads\ulb3.jpg',
               
               r'C:\Users\student\Downloads\ulb4.jpg',
               
               r'C:\Users\student\Downloads\ulb5.jpg']
```

```python
for image_path in image_paths :

    img = cv2.imread(image_path)

    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    edges = cv2.Canny(gray, 100, 200)
    
    plt.subplot(121), plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)), plt.title('Original Image')
    
    plt.subplot(122), plt.imshow(edges, cmap='gray'), plt.title('Edge Image')
    
    plt.show()
```
